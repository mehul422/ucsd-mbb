```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import re

# Path to YOLOv8 label .txt files
label_path = "/Users/mehulverma422/Desktop/eclipse/ucsd-mbb/data/runs/detect/duke_inference/labels"

# Video resolution used during detection
img_width, img_height = 1920, 1080

# Convert all YOLO .txt labels into a DataFrame
data = []

for filename in sorted(os.listdir(label_path)):
    if filename.endswith(".txt"):
        match = re.search(r"(\d+)", filename)
        if match:
            frame_num = int(match.group(1))
        else:
            continue  # skip if no number found

        with open(os.path.join(label_path, filename), "r") as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 5:
                    class_id, x_center, y_center, w, h = map(float, parts[:5])
                else:
                    continue  # skip malformed lines

                # Convert normalized YOLO coords to pixel coords
                x = x_center * img_width
                y = y_center * img_height

                data.append({
                    "Frame": frame_num,
                    "ClassID": int(class_id),
                    "X": round(x),
                    "Y": round(y),
                    "W": round(w * img_width),
                    "H": round(h * img_height)
                })


df = pd.DataFrame(data)
# Sort + compute bounding box area
df = df.sort_values("Frame").reset_index(drop=True)
df["Area"] = df["W"] * df["H"]

# Optional: Filter out boxes that are way too big or small (noise)
df = df[(df["Area"] > 100) & (df["Area"] < 3000)]

# âœ… Keep only the largest box per frame (best guess for ball)
df_top_ball = (
    df.sort_values("Area", ascending=False)
      .groupby("Frame")
      .head(1)
      .reset_index(drop=True)
)

# âœ… Score each remaining ball detection based on Y position
def score_ball_position(y):
    if y > 900:
        return 100
    elif y > 700:
        return 80
    elif y > 500:
        return 60
    elif y > 300:
        return 40
    else:
        return 20

df_top_ball["ShotQuality"] = df_top_ball["Y"].apply(score_ball_position)

plt.figure(figsize=(12, 4))
plt.scatter(df_top_ball["Frame"], df_top_ball["ShotQuality"], color="green", s=10)
plt.title("Basketball Shot Quality Over Time (1 Detection Per Frame)")
plt.xlabel("Frame")
plt.ylabel("Score (0â€“100)")
plt.axhline(70, linestyle="--", color="gray", alpha=0.5, label="Good shot threshold")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# NBA court size in feet
court_length = 94
court_width = 50

# Assuming 1920x1080 pixel resolution â†’ scale to feet
df_top_ball["X_ft"] = df_top_ball["X"] / 1920 * court_length
df_top_ball["Y_ft"] = df_top_ball["Y"] / 1080 * court_width

plt.figure(figsize=(16, 8))

# Draw court boundaries
plt.plot([0, 94], [0, 0], 'k-')     # Baseline
plt.plot([0, 94], [50, 50], 'k-')   # Top line
plt.plot([0, 0], [0, 50], 'k-')     # Left sideline
plt.plot([94, 94], [0, 50], 'k-')   # Right sideline

# Plot ball movement
plt.scatter(df_top_ball["X_ft"], df_top_ball["Y_ft"], c=df_top_ball["Frame"], cmap='viridis', s=10)
plt.colorbar(label="Frame #")
plt.title("Basketball Movement Across Full Court")
plt.xlabel("Court Length (ft)")
plt.ylabel("Court Width (ft)")
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
# Calculate frame-to-frame movement
df_top_ball["dx"] = df_top_ball["X_ft"].diff()
df_top_ball["dy"] = df_top_ball["Y_ft"].diff()
df_top_ball["dist"] = np.sqrt(df_top_ball["dx"]**2 + df_top_ball["dy"]**2)

# Detect direction change (sign flip)
df_top_ball["dir_flip"] = np.sign(df_top_ball["dx"]) != np.sign(df_top_ball["dx"].shift(1))

# Possession change: large direction reversal
df_top_ball["PossessionChange"] = df_top_ball["dir_flip"] & (df_top_ball["dist"] > 5)

# Shot attempts: spike in ShotQuality to 100
df_top_ball["ShotAttempt"] = (df_top_ball["ShotQuality"] == 100) & (df_top_ball["ShotQuality"].shift(1) < 100)

# Fast break: rapid movement (>10 ft between frames)
df_top_ball["FastBreak"] = df_top_ball["dist"] > 10

plt.figure(figsize=(16, 8))

# Court outline
plt.plot([0, 94], [0, 0], 'k-')
plt.plot([0, 94], [50, 50], 'k-')
plt.plot([0, 0], [0, 50], 'k-')
plt.plot([94, 94], [0, 50], 'k-')

# Plot full path
plt.plot(df_top_ball["X_ft"], df_top_ball["Y_ft"], color='gray', linewidth=1, alpha=0.4)

# Mark events
plt.scatter(df_top_ball.loc[df_top_ball["PossessionChange"], "X_ft"],
            df_top_ball.loc[df_top_ball["PossessionChange"], "Y_ft"],
            color="orange", label="Possession Change", s=60)

plt.scatter(
    df_top_ball.loc[df_top_ball["ShotAttempt"], "X_ft"],
    df_top_ball.loc[df_top_ball["ShotAttempt"], "Y_ft"],
    color="red",
    label="Shot Attempt",
    s=100,           # Make the dots larger
    edgecolors="black",  # Optional: outline for visibility
    zorder=10        # Draw on top of everything else
)

plt.scatter(df_top_ball.loc[df_top_ball["FastBreak"], "X_ft"],
            df_top_ball.loc[df_top_ball["FastBreak"], "Y_ft"],
            color="blue", label="Fast Break", s=60)

plt.title("Event-Tagged Ball Movement on Full Court")
plt.xlabel("Court Length (ft)")
plt.ylabel("Court Width (ft)")
plt.legend()
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


df_top_ball["ShotSuccess"] = False

# If no major movement after a 100-score shot, mark as success
for i in df_top_ball[df_top_ball["ShotAttempt"]].index:
    y_post = df_top_ball.loc[i:i+5, "Y_ft"].values
    if len(y_post) > 1 and all(abs(np.diff(y_post)) < 2):  # minimal bounce
        df_top_ball.loc[i, "ShotSuccess"] = True

features = df_top_ball[df_top_ball["ShotAttempt"]].copy()

# If Y_ft is very close to hoop and minimal vertical change after the shot, assume made
features["ShotSuccess"] = (
    (features["ShotQuality"] == 100) &
    (features["Y_ft"].diff().abs() < 1)  # minimal ball bounce
)

print(features["ShotSuccess"].value_counts())

features["CourtSide"] = features["X_ft"].apply(lambda x: "Left" if x < 47 else "Right")
features["DistanceFromBaseline"] = 50 - features["Y_ft"]  # Assuming basket near baseline

# Add movement-based features
features["SpeedBefore"] = df_top_ball["dist"].shift(1).fillna(0).loc[features.index]
features["SpeedAfter"] = df_top_ball["dist"].shift(-1).fillna(0).loc[features.index]

# Select final features
X = features[["X_ft", "Y_ft", "DistanceFromBaseline", "SpeedBefore", "SpeedAfter"]]
y = features["ShotSuccess"]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

features["PredictedSuccess"] = model.predict(X)

plt.figure(figsize=(16, 8))
for success, color in zip([True, False], ["green", "red"]):
    subset = features[features["PredictedSuccess"] == success]
    plt.scatter(subset["X_ft"], subset["Y_ft"], color=color, label=f"Predicted {'Make' if success else 'Miss'}", s=100, alpha=0.7)

plt.title("Shot Success Prediction (Logistic Regression)")
plt.xlabel("Court Length (ft)")
plt.ylabel("Court Width (ft)")
plt.legend()
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```
